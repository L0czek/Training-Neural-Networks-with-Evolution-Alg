```python
n = ['DiffEvo-Poly-0.5', 'DiffEvo-SinCos-0.1', 'DiffEvo-Xor-0.1', 
     'MuLambda-Poly-0.5-0.6', 'MuLambda-SinCos-0.9-1.0', 'MuLambda-Xor-1.3-1.5',
     'DGradientDescent-Poly-1e-3-40', 'DGradientDescent-SinCos-1e-3-200', 
     'DGradientDescent-Xor-1e-3-200' ]
np = n[0::3]
ns = n[1::3]
nx = n[2::3]
```

```python
def plot(exps, title):
    for name in exps:
        with open('results/' + name + '.pickle', "rb") as file:
            experiment = pickle.load(file)
            Y = [ experiment.best_individual_loss ]
            X = [ experiment.experiment_time_in_seconds ]
            plt.scatter(X, Y, label=name)
    plt.legend()
    plt.xlabel('Czas wykonania w sekundach [s]')
    plt.ylabel('Wartość funkcji straty najlepszej sieci')
    plt.grid('on')
    plt.title(title)
    plt.subplots_adjust(left=0.1)
    plt.savefig(title + '.png', format='png', bbox_inches="tight")
    plt.show()
```

```python
plot(np, 'Porównanie czasu wykonania do jakości dla wielomianów')
```

```python
plot(ns, 'Porównanie czasu wykonania do jakości dla sinusa')
```

```python
plot(nx, 'Porównanie czasu wykonania do jakości dla xor\'a')
```
